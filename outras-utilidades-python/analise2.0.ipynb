```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
from importlib import reload
import numpy as np
import csv

plt = reload(plt)
```

# CODE


```python
FILE_REFACTORIN = "refactorings-auto.csv"
FILE_COMMENTS = "comments-auto.csv"
FILE_NAME_TABELA_DIFERENCA_COMENTARIOS = "tabela-diferenca-comentarios-auto.csv"
FILE_NAME_TABELA_MERGE = "tabela-merge-auto.csv"
OUTPUT = "analise-auto-refatorado-auto.csv"
```

```python
df_refactorings = pd.read_csv(FILE_REFACTORIN)
df_refactorings.drop(
    columns=['12_MOVEDTOPATHCLASS', '14_OCORRIDO'], inplace=True)

DF_COMMENTS = pd.read_csv(FILE_COMMENTS)
DF_COMMENTS.drop(columns=['10_HASH', '13_TYPE',
                 '14_STARTLINE', '15_ENDLINE'], inplace=True)
```

```python
agrupado = DF_COMMENTS.groupby(
    ["11_HASH_CLASSPATH", "12_PARENTHASH_CLASSPATH"]).sum().reset_index()

# Renomeando as colunas para os novos nomes
agrupado.rename(columns={
    "16_SEGMENTOS": "QNT_SEGMENTOS",
    "18_ISBLOCKCOMMENT": "QNT_ISBLOCKCOMMENT",
    "19_ISLINECOMMENT": "QTN_ISLINECOMMENT",
    "20_ISJAVADOCCOMMENT": "QTN_ISJAVADOCCOMMENT",
    "21_ISORPHANCOMMENTA": "QTN_ISORPHANCOMMENTA"

}, inplace=True)


merged_df = pd.merge(agrupado, agrupado, right_on="11_HASH_CLASSPATH",
                     left_on="12_PARENTHASH_CLASSPATH", suffixes=('', '_PARENTE'))

merged_df.drop(columns=['11_HASH_CLASSPATH_PARENTE',
               '12_PARENTHASH_CLASSPATH_PARENTE'], inplace=True)

merged_df.to_csv(FILE_NAME_TABELA_MERGE, index=False)
```

```python

merged_df["DIFERENCA_QNT_ISBLOCKCOMMENT"] = merged_df["QNT_ISBLOCKCOMMENT"] - \
    merged_df["QNT_ISBLOCKCOMMENT_PARENTE"]
merged_df["DIFERENCA_QTN_ISLINECOMMENT"] = merged_df["QTN_ISLINECOMMENT"] - \
    merged_df["QTN_ISLINECOMMENT_PARENTE"]
merged_df["DIFERENCA_QTN_ISJAVADOCCOMMENT"] = merged_df["QTN_ISJAVADOCCOMMENT"] - \
    merged_df["QTN_ISJAVADOCCOMMENT_PARENTE"]
merged_df["DIFERENCA_QTN_ISORPHANCOMMENTA"] = merged_df["QTN_ISORPHANCOMMENTA"] - \
    merged_df["QTN_ISORPHANCOMMENTA_PARENTE"]
merged_df["DIFERENCA_QNT_SEGMENTOS"] = merged_df["QNT_SEGMENTOS"] - \
    merged_df["QNT_SEGMENTOS_PARENTE"]
```

```python
novo_df_com_diferencas = merged_df[[
    "11_HASH_CLASSPATH", "12_PARENTHASH_CLASSPATH",
    "QNT_ISBLOCKCOMMENT", "QTN_ISLINECOMMENT", "QTN_ISJAVADOCCOMMENT", "QTN_ISORPHANCOMMENTA",
    "DIFERENCA_QNT_ISBLOCKCOMMENT", "DIFERENCA_QTN_ISLINECOMMENT", "DIFERENCA_QNT_SEGMENTOS",
    "DIFERENCA_QTN_ISJAVADOCCOMMENT", "DIFERENCA_QTN_ISORPHANCOMMENTA"
]]


novo_df_com_diferencas.drop(columns=['12_PARENTHASH_CLASSPATH', 'QNT_ISBLOCKCOMMENT',  'QTN_ISLINECOMMENT', 'QTN_ISJAVADOCCOMMENT',
                            'QTN_ISORPHANCOMMENTA', 'DIFERENCA_QTN_ISORPHANCOMMENTA', 'DIFERENCA_QNT_SEGMENTOS'], inplace=True)
novo_df_com_diferencas.to_csv(
    FILE_NAME_TABELA_DIFERENCA_COMENTARIOS, index=False)
novo_df_com_diferencas
```

```python
merged_df = pd.merge(
    df_refactorings, novo_df_com_diferencas, on="11_HASH_CLASSPATH")

merged_df
```

```python

# Transformando a coluna '13_REFACTORINGTYPE' em códigos numéricos
merged_df['REFACTORING_CODE'] = merged_df['13_REFACTORINGTYPE'].astype(
    'category').cat.codes


# Selecionando apenas colunas numéricas para calcular a correlação
numeric_cols = merged_df.select_dtypes(include=['number'])

# Calculando a correlação entre as colunas numéricas
correlations = numeric_cols.corr()

# Exibindo as correlações relevantes
print(correlations[['REFACTORING_CODE', 'DIFERENCA_QNT_ISBLOCKCOMMENT',
      'DIFERENCA_QTN_ISLINECOMMENT', 'DIFERENCA_QTN_ISJAVADOCCOMMENT']])
```

```python
merged_df['HAS_REFACTORING'] = merged_df['13_REFACTORINGTYPE'].apply(
    lambda x: 1 if x != "" else 0)

merged_df_copy = merged_df.copy()
# Transformando as colunas de mudanças de comentários em binário (0 ou 1)

# Considerando que qualquer valor diferente de 0 indica que houve uma mudança

merged_df['HAS_BLOCKCOMMENT_CHANGE'] = merged_df['DIFERENCA_QNT_ISBLOCKCOMMENT'].apply(
    lambda x: 0 if x == 0 else 1)

merged_df['HAS_LINECOMMENT_CHANGE'] = merged_df['DIFERENCA_QTN_ISLINECOMMENT'].apply(
    lambda x: 0 if x == 0 else 1)

merged_df['HAS_JAVADOCCOMMENT_CHANGE'] = merged_df['DIFERENCA_QTN_ISJAVADOCCOMMENT'].apply(
    lambda x: 0 if x == 0 else 1)


# Mantendo apenas as colunas relevantes para a análise de associação

df_for_association = merged_df[['HAS_REFACTORING', 'HAS_BLOCKCOMMENT_CHANGE',
                                'HAS_LINECOMMENT_CHANGE', 'HAS_JAVADOCCOMMENT_CHANGE']]


# Aplicando o algoritmo Apriori para encontrar itemsets frequentes

frequent_itemsets = apriori(
    df_for_association, min_support=0.01, use_colnames=True)


# Gerando as regras de associação

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)


# Exibindo as regras
rules
```
